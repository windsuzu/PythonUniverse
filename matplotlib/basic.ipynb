# Matplotlib Basic

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
# with this magic function, we can skip `plt.show()`
%matplotlib inline
````

## Intro to Pyplot

### Plotting y-data only

````python
plt.plot(np.arange(100)**2)
````

### Plotting single x-y data

````python
x = np.array([1, 2, 3, 4])
y = x**2
plt.plot(x, y, marker="o")
````

### Plotting multiple x-y data pairs

````python
x1, y1 = range(100), np.random.default_rng(seed=1).normal(0, 0.5, 100)  # mean, std, size
x2, y2 = range(100), np.random.default_rng(seed=2).normal(1, 1, 100)
x3, y3 = range(100), np.random.default_rng(seed=3).normal(5, 0.5, 100)
````

````python
plt.plot(x1, y1, x2, y2, x3, y3)
````

## Add Annotations

We can decorate our plot (`Line2D`) with tons of properties in `pyplot` and `pyplot.plot()`:

### `pyplot.plot()` Parameters

- `marker`
- `markersize`
- `linestyle` or `ls`
- `linewidth` or `lw`
- `color` or `c`
- `label` (required to apply `pyplot.legend()`)
- `alpha`
- `visible`
- ...

In `pyplot.plot()`, you can use `format strings` as a shortcut to set colors, markers and lines.

``` py
fmt = '[marker][line][color]'

# For example, the following two settings are the same.

plt.plot(x, y, "o--r")
plt.plot(x, y, marker="o", ls="--" c="r")
```

-  To see all available `marker`, `color`, `line` types, please visit the [documentation](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html).

````python
plt.plot(np.sin(np.linspace(0, 10, 100)), "*-b", lw=2, markersize=5, label="sin(x)")
plt.plot(np.cos(np.linspace(0, 10, 100)), c="g", ls="--", marker=".", lw=2, markersize=5, label="cos(x)")
````

### `pyplot` Functions

- `pyplot.xlabel`
- `pyplot.ylabel`
- `pyplot.title`
- `pyplot.grid`
- `pyplot.legend`
- `pyplot.text`
- `pyplot.annotate`
- ...

````python
plt.plot(np.sin(np.linspace(0, 10, 100)), label="sin(x)")
plt.plot(range(100), [0]*100, label="0")
plt.xlabel("X here")
plt.ylabel("Y here")

plt.title("sin(x) and 0")

plt.grid()

plt.legend()

plt.text(x=70, y=-1, s="hahahaha")

plt.annotate("wow \nmax", xy=(16, 1), xytext=(40, 0.9), arrowprops={"facecolor": "orange", "shrink": 0.05})
plt.annotate("wow \nmax again", xy=(78, 1), xytext=(95, 0.9), arrowprops={"facecolor": "red", "shrink": 0.05})
````
